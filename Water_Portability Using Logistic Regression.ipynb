```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE
```

```python
df=pd.read_csv("water_potability.csv")
df
```

```python
df.isnull().sum()
```

```python
df.mean()
```

```python
df.fillna(df.mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['Potability'].value_counts()
```

```python
X=df.drop('Potability',axis='columns')
X
```

```python
y=df['Potability']
y
```

```python
smote=SMOTE(sampling_strategy='minority')
XX,yy=smote.fit_resample(X,y)
```

```python
yy.value_counts()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(XX,yy,random_state=0,test_size=0.20,stratify=yy)
```

```python
ytrain.value_counts(),ytest.value_counts()
```

```python
xtrain
```

```python
xtest
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(xtrain,ytrain)
```

```python
lr.predict([[6.448931033,240.244771,13979.16572,9.077984756,314.5904772,473.7513016,17.41690264,84.0247864,3.622195522]])
```

```python
ypred=lr.predict(xtest)
ypred
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(ytest,ypred)
ac*100
```

```python
import pickle
pickle.dump(lr, open('model.pkl', 'wb'))
```
